```python
import pickle as pkl

data = pkl.load(open('diversevul_graph_info.pkl','rb'))
data
```

```python
data2 = pkl.load(open('cross_val_5folds_train_val_notestfiles_with_files_not_inprocessedwdeg.pkl','rb'))
```

```python
# inverse mapping 
hash_to_cwe = {}
for d in data:
    hash_to_cwe[str(d['hash'])] = d['cwe']
```

```python
cwe_counts = {} 
for i, fold in enumerate(data2):
    
    for file in fold:
        cwes = hash_to_cwe[file.split('_')[1]]
        for cwe in cwes:
            if cwe not in cwe_counts:
                cwe_counts[cwe] = [0,0,0,0,0] 
            cwe_counts[cwe][i] += 1
        
    
```

```python
# get cwe ratio
cwe_ratios = {}
for cwe, l in cwe_counts.items():
    ratio = []
    for i in range(len(l)):
        ratio.append(l[i]/sum(l))
    
    # print with .2f precision
    ratio = [round(r,2) for r in ratio]
    print(cwe, ratio)
```
